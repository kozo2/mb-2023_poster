```python
import json
# write dictionary values to a file as a tab-delimited table
with open('ST_metabolites_all.tsv', 'w') as f:
    for jsonpath in stmetabolites:
        with open(jsonpath) as f2:
            data = json.load(f2)
            if type(data) is list:
                pass
            else:
                #print(jsonpath)
                keys = data.keys()
                if 'refmet_name' in keys:
                    f.write(data['study_id'] + '\t')
                    f.write(data['analysis_id'] + '\t')
                    f.write(data['analysis_summary'] + '\t')
                    f.write(data['refmet_name'] + '\n')
                else:
                    for key in keys:
                        metabolite = data[key]
                        if metabolite['refmet_name'] != "":
                            f.write(metabolite['study_id'] + '\t')
                            f.write(metabolite['analysis_id'] + '\t')
                            f.write(metabolite['analysis_summary'] + '\t')
                            f.write(metabolite['refmet_name'] + '\n')
```
