```python
# Create a tsv file with molfile path and the inchikey
with open('KEGG_compoundID_inchikey.tsv', 'w') as f_output:
    for molfile_path in molfilepaths:
        print(molfile_path)
        mol = Chem.MolFromMolFile(molfile_path)
        inchikey = inchi.MolToInchiKey(mol)
        KEGG_compoundID = molfile_path.split('\\')[-1].split('.')[0]
        f_output.write(KEGG_compoundID + '\t' + inchikey + '\n')
```
